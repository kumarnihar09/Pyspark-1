<b>How many employees are in the input file?

```
# Write your code using the Spark Function
df = spark.read.csv("dbfs:/FileStore/tables/EmployeeData_8xfZvOlzqL.csv", header=True, inferSchema=True)
##df.show()
##default separator is comma
#display(df)
df.count()
```

<b> What is the average salary of all the employees?

```
# Write your code using the Spark Function
from pyspark.sql.functions import *
df1 = df.select(avg(df.salary).alias('Avg_salary'))
display(df1)
```

<b> How many male and female employees are there in the input file?

```
# Write your code using the Spark Function
#Trim down data first then do aggregation
#df2 = df.filter("gender is not null").groupBy(df.gender).count().collect()
df2 = df.filter("gender is not null").groupBy(df.gender).agg(count(df.gender).alias('count'),max(df.salary).alias('max_salary'))
display(df2)
```

<b> Who is the employee with the highest salary?

```
# Write your code using the Spark Function
from pyspark.sql.window import *
##w = Window().partitionBy(df.gender).orderBy(df.salary.desc_nulls_last())
##df3 = df.withColumn('row_num', row_number().over(w))
##display(df3.filter("row_num == 1"))
w = Window().orderBy(df.salary.desc_nulls_last())
df3 = df.withColumn('row_num', row_number().over(w))
display(df3.filter("row_num == 1"))
```

<b> What is the total bonus amount paid to all the employees?

```
# Write your code using the Spark Function
df4 = df.select(sum('bonus').alias('Total_Bonus'))
display(df4)
```

<b> Who are the top 5 employees with the highest salaries?

```
# Write your code using the Spark Function
w = Window().orderBy(df.salary.desc_nulls_last())
df6 = df.withColumn('row_num', row_number().over(w))
display(df6.filter("row_num <= 5"))
```
