<b> How many employees have a first name that starts with the letter 'J'?

```
# Write your code using the Spark Function
from pyspark.sql.functions import *
df = spark.read.csv('dbfs:/FileStore/EmployeeData_HhTLiz9LdC.csv', header='True',inferSchema='True')
df.filter(df.first_name.startswith('J')).count()
```

d
<b> How many employees have a last name that ends with the letter 's'?

```
# Write your code using the Spark Function
df.filter(df.last_name.endswith('s')).count()
```

d
<b> What is the length of the longest first name?

```
# Write your code using the Spark Function
df2 = df.withColumn('len_longest_first', length(df.first_name))
df2.agg(max(df2.len_longest_first)).show()
```

d
<b> What is the length of the shortest last name?

```
# Write your code using the Spark Function
df3 = df.withColumn('len_shortest_last', length(df.last_name))
df3.agg(min(df3.len_shortest_last)).show()
#min_salary = df.agg(min(df.salary).alias('min')).collect()[0].min
#df.filter(df.salary == min_salary).show()
#collect always returns a list
```

d
<b> What is the email domain of the employee with the highest salary?

```
# Write your code using the Spark Function
max_salary = df.agg(max(df.salary).alias('max')).collect()[0].max
df.filter(df.salary == max_salary).select(split(df.email,'@')[1].alias('domain'),df.first_name).show()
```

<b> What is the first name of the employee with the email address that contains the substring 'hr'?

```
# Write your code using the Spark Function
df.filter(df.email.contains('hr')).show()
```

d
<b> What is the last name of the employee with the email address that ends with the domain '.com'?

```
# Write your code using the Spark Function
df.filter(df.email.endswith('.com')).select(df.last_name,df.email).show()
```

d
<b> How many employees have an email address that contains their last name?

```
# Write your code using the Spark Function
df.filter(df.email.contains(lower(df.last_name))).count()
```

d
<b> What is the email of the employee with the first name that contains a space character?

```
# Write your code using the Spark Function
df.filter(df.first_name.contains(' ')).select(df.email).show()
```

d
<b> What is the email of the employee with the last name that contains a hyphen character?

```
# Write your code using the Spark Function
df.filter(df.last_name.contains('-')).select(df.last_name,df.email).show()
```
